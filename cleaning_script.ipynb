```python
import pandas as pd
```

```python
# Step 1: Load the dataset
sales = pd.read_csv("/content/supermarket_sales - Sheet1.csv")
```

```python
sales.head()
```

```python
# To check number of rows and columns
sales.shape
```

```python
sales.columns
```

```python
# Step 2: Remove duplicate rows
sales = sales.drop_duplicates()
```

```python
sales.dtypes
```

```python
#step 3: Convert 'Date' to datetime and 'time' to time object
sales['date'] = pd.to_datetime(sales['date'], format='%m/%d/%Y', errors='coerce')  # Specifying format explicitly. Changed format to handle "1/27/2019" like inputs
# Applying strip to remove leading/trailing spaces and then converting to datetime
# Convert 'time' column to string before using .str.strip()
sales['time'] = pd.to_datetime(sales['time'].astype(str).str.strip(), format='%H:%M', errors='coerce').dt.time # Changed 'Time' to 'time'
```

```python
#step 4: Standardize column names (lowercase, no spaces)
sales.columns = sales.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
#step 5: Standardize categorical values
sales['gender'] = sales['gender'].str.capitalize()
sales['customer_type'] = sales['customer_type'].str.capitalize()
sales['payment'] = sales['payment'].str.title()
```

```python
# Step 6: Export the cleaned dataset
sales.to_csv("cleaned_supermarket_sales.csv", index=False)
```
